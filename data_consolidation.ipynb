```python
# we code a brief AI scrypt that consolidates our csv files so they can be pushed to git
import pandas as pd

def sample_csv(input_path, output_path, n=50000, random_state=42):
    df = pd.read_csv(input_path)
    df.sample(n=n, random_state=random_state).to_csv(output_path, index=False)
```
